```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv(r'C:\Users\Stayt\dss_table_wo_2018.csv', sep=';')

#df.head()

#df.isnull().any()

#df.shape

#df.info()

#summary(df)

#Sales_out = df['SOutChain_FixPrice']
Sales_out = df['SOutChain']

pd.to_numeric(Sales_out, errors='coerce')

df['SOutChain_FixPrice'] = Sales_out.values
    

X = df[['Category_D_or_K', 'Trax_active', 'Photoban', 'Trax_quarter_active', 'Trax_introduction']].values
y = Sales_out.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


regressor = LinearRegression()  
regressor.fit(X, y)

intercept = regressor.intercept_
coefficients = regressor.coef_

y_pred = regressor.predict(X_test)

# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))


columns = ["Variables", 'Coefficients']
data = (
    ['Intercept', intercept], 
    ['Category D or K', coefficients[0]], 
    ['Trax active', coefficients[1]], 
    ['Ban on photo', coefficients[2]], 
    ['Trax Active for 3 months', coefficients[3]], 
    ['Introduction of Trax', coefficients[4]]
       )

representation_table = pd.DataFrame(data=data, columns=columns)

print(representation_table)


#sns.kdeplot(y[:48], shade=True)

#ban_on_photo = df[df["Photoban"]==1]

#ban_on_photo.describe()
```
